# Data Preprocessing

```python
# importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# read csv file to dataframe
df = pd.read_csv("C:/Users/chino/OneDrive/Desktop/DataSci_folder/DataScienceProjects_Mentorship/MY DATASETS/Telco-Customer-Churn.csv")
# set index as CustomerID
df = df.set_index('customerID')
print(df.shape)
df.head()
```

```python
df.info()
```

```python
# correcting mismatching datatypes

# convert 'TotalCharges' values to numeric values
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# check for missing values after conversion
# missing_value = df['TotalCharges'].isna().sum()
# print(f"The number of missing values is {missing_value}.")   the number of missing values was 11

# fill missing values with the median
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())

# check missing values
missing_value = df['TotalCharges'].isna().sum()
print(f"The number of missing values is {missing_value}.")

# check datatype 
print(df['TotalCharges'].dtype)
```

# Exploratory Data Analysis

```python
# Customer churn rate by demographics
demo = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']

for col in demo:
    # established canva
    plt.figure(figsize=(6,4))
    
    # created churn rate feature
    churn_rate_df = (
        df.groupby(col)['Churn']
          .apply(lambda x: (x == 'Yes').mean())
          .reset_index(name='Churn_Rate')
    )
    # plotting
    sns.barplot(
        data=churn_rate_df,
        x=col,
        y='Churn_Rate',
        palette=['#800020', '#1f77b4']
    )

    plt.title(f'Customer Churn Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('Churn Rate')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# customer churn rate by service
services = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
            'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
            'Contract', 'PaperlessBilling', 'PaymentMethod']

for col in services:
    plt.figure(figsize=(6,4))

    churn_rate_df = (
        df.groupby(col)['Churn']
          .apply(lambda x: (x == 'Yes').mean())
          .reset_index(name='Churn_Rate')
    )

    sns.barplot(
        data=churn_rate_df,
        x=col,
        y='Churn_Rate',
        palette=['#800020', '#1f77b4']
    )

    plt.title(f'Customer Churn Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('Churn Rate')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# create tenure bins
tenure_bins = [0, 12, 24, 36, 48, 60, 72]
tenure_labels = ['01 yr', '12 yrs', '23 yrs', '34 yrs', '45 yrs', '5+ yrs']

# create Tenure_Group feature
df['Tenure_Group'] = pd.cut(
    df['tenure'],
    bins=tenure_bins,
    labels=tenure_labels,
    right=False
)
```

```python
# churn rate by Tenure

# draw canvas
plt.figure(figsize=(7,4))

# tenure churn rate df
tenure_churn_rate = (
    df.groupby('Tenure_Group')['Churn']
      .apply(lambda x: (x == 'Yes').mean())
      .reset_index(name='Churn_Rate')
)
# plotting
sns.barplot(
    data=tenure_churn_rate,
    x='Tenure_Group',
    y='Churn_Rate',
    palette=['#800020', '#1f77b4']
)

plt.title('Customer Churn Rate by Tenure')
plt.xlabel('Tenure Group')
plt.ylabel('Churn Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## INSIGHT:

From the BarChart above Users of 0-1 year using the service tend to churn the highest... Having 48% churn rate.

```python
# calculate the mrr_churn_rate
mrr_churn_rate = df.loc[df['Churn'] == 'Yes', 'MonthlyCharges'].sum() / df['MonthlyCharges'].sum()
mrr_churn_rate
```

```python
# MRR churn rate by Demo (no warning, future-proof)
demo = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']

for col in demo:
    plt.figure(figsize=(6,4))

    mrr_churn_df = (
        df.assign(
            Churned_MRR = df['MonthlyCharges'].where(df['Churn'] == 'Yes', 0)
        )
        .groupby(col)
        .agg(
            Total_MRR = ('MonthlyCharges', 'sum'),
            Churned_MRR = ('Churned_MRR', 'sum')
        )
        .reset_index()
    )

    mrr_churn_df['MRR_Churn_Rate'] = mrr_churn_df['Churned_MRR'] / mrr_churn_df['Total_MRR']

    sns.barplot(
        data=mrr_churn_df,
        x=col,
        y='MRR_Churn_Rate',
        palette=['#800020', '#1f77b4']
    )

    plt.title(f'MRR Churn Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('MRR Churn Rate')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# MRR Churn rate by services
services = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
            'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
            'Contract', 'PaperlessBilling', 'PaymentMethod']

for col in services:
    plt.figure(figsize=(6,4))

    mrr_churn_df = (
        df.assign(
            Churned_MRR = df['MonthlyCharges'].where(df['Churn'] == 'Yes', 0)
        )
        .groupby(col)
        .agg(
            Total_MRR = ('MonthlyCharges', 'sum'),
            Churned_MRR = ('Churned_MRR', 'sum')
        )
        .reset_index()
    )
    
    mrr_churn_df['MRR_Churn_Rate'] = mrr_churn_df['Churned_MRR'] / mrr_churn_df['Total_MRR']
    
    sns.barplot(
        data=mrr_churn_df,
        x=col,
        y='MRR_Churn_Rate',
        palette=['#800020', '#1f77b4']
    )

    plt.title(f'MRR Churn Rate by {col}')
    plt.xlabel(col)
    plt.ylabel('MRR Churn Rate')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# draw canvas
plt.figure(figsize=(7,4))

tenure_mrr_churn = (
    df.assign(
        Churned_MRR = df['MonthlyCharges'].where(df['Churn'] == 'Yes', 0)
    )
    .groupby('Tenure_Group')
    .agg(
        Total_MRR = ('MonthlyCharges', 'sum'),
        Churned_MRR = ('Churned_MRR', 'sum')
    )
    .reset_index()
)

tenure_mrr_churn['MRR_Churn_Rate'] = tenure_mrr_churn['Churned_MRR'] / tenure_mrr_churn['Total_MRR']

sns.barplot(
    data=tenure_mrr_churn,
    x='Tenure_Group',
    y='MRR_Churn_Rate',
    palette=['#800020', '#1f77b4']
)

plt.title('MRR Churn Rate by Tenure')
plt.xlabel('Tenure Group')
plt.ylabel('MRR Churn Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Customer Churn & MRR Churn Analysis
## Business Objective

Reduce customer churn and revenue leakage (MRR churn) by identifying high-risk segments and the service factors driving attrition.

Key KPIs Tracked

Customer Churn Rate (%) i.e % of customers who leave

MRR Churn Rate (%) i % of monthly revenue lost due to churn

Tenure (01 year focus)  early lifecycle risk indicator

These KPIs directly reflect:

Customer retention health

Revenue stability

Effectiveness of onboarding & service quality

## Demographic Insights  Who Is Leaving?
### Gender

Customer Churn: Male 28% | Female 26%

MRR Churn: Male 30% | Female 32%
Insight: Gender is not a strong churn driver. Churn is behavior-driven, not demographic here.

Business Implication:
Do not segment or target by gender. Focus resources elsewhere.

### Senior Citizens

Customer Churn: Senior 42% | Non-Senior 24%

MRR Churn: Senior 43% | Non-Senior 27%

#### Insight: Senior citizens are high-risk and high-revenue-loss customers.

##### Business Problem:
Older customers may be struggling with:

Service complexity

Technical setup

Poor support experience

##### Actionable Solutions:

Dedicated Senior Support Line

Simplified plans & bundles

Proactive check-in calls after onboarding

## Partner & Dependents

Without Partner: 33% churn | 19% MRR churn

With Partner: 20% churn | 27% MRR churn

Non-Dependent: 33% churn | 35% MRR churn

Dependent: 17% churn | 19% MRR churn

#### Insight:

Single & non-dependent customers churn more

But customers with partners contribute more to MRR churn when they leave

##### Business Problem:
Single customers are less sticky. Partnered customers are higher value but vulnerable when dissatisfied.

### Actionable Solutions:

Family bundles & multi-line discounts

Household loyalty pricing

Couple/family retention offers before contract end

## Service & Product Insights  Why Are They Leaving?

### Internet Service (Biggest Red Flag)

Fiber Optic: 44% churn | 41% MRR churn

DSL: 19% churn | 16% MRR churn

No Internet: ~8% churn & MRR churn

#### Insight:
Fiber customers are your highest churn AND highest revenue loss segment.

##### Business Problem:
This usually indicates:

Poor network stability

Speed inconsistency

Pricing dissatisfaction

#### Actionable Solutions:

Immediate Fiber service quality audit

Region-based performance analysis

Compensation/discount campaigns for unstable zones

## Add-On Services (OnlineSecurity, Backup, TechSupport, DeviceProtection)

Consistently:

With Add-ons: ~1525% churn

Without Add-ons: ~3943% churn

#### Insight:
Customers without add-ons churn nearly 23x more.

Business Meaning:
Add-ons increase stickiness and perceived value.

#### Actionable Solutions:

Bundle add-ons into onboarding offers

Free trial of security/backup for first 60 days

Protection Pack bundles

## Contract Type (Very Strong Signal)

Monthly: 43% churn | 49% MRR churn

Yearly: 12% churn | 15% MRR churn

Two-Year: 3% churn | 4% MRR churn

#### Insight:
Monthly contracts are your biggest leakage point.

Business Problem:
Low commitment = easy exit.

#### Actionable Solutions:

Aggressive monthly  yearly upgrade campaigns

Loyalty discounts at month 23

Contract upgrade prompts before month 6

## Payment Method

Electronic Check: 44% churn | 49% MRR churn

Others (Bank, Card, Mail): ~1728%

#### Insight:
Electronic Check users are high risk & high revenue loss.

Likely Causes:

Friction in payment process

Failed payments

Lower engagement

#### Actionable Solutions:

Incentivize auto-pay (discounts, data bonuses)

Payment experience redesign

Proactive reminders & retry logic

## Tenure Analysis  When Do They Leave?
01 Year Customers

Customer Churn: 48%

MRR Churn: 57%

#### Insight:
The first year is the danger zone.

Business Problem:
Onboarding, expectation setting, and early experience are failing.

#### Actionable Solutions:

Structured onboarding journey

First 90-day check-in campaign

Usage education emails & in-app tips

Early loyalty rewards

## In summary
Our churn is not random. It is highly concentrated among early-tenure customers, Fiber Optic users, monthly contract customers, and those without value-added services. These same segments are also responsible for the highest revenue loss. By improving onboarding, strengthening service quality, and converting customers to longer contracts with bundled services, we can significantly reduce both customer churn and revenue churn
